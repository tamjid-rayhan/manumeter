```python
import pandas as pd
import numpy as np
```

```python
def read_matfile_to_pd(file, feature_list):
    '''
    This function returns a pandas DataFrame from a given
    .mat file with columns from a provided feature list.
    '''
    #importing necessary library
    from scipy.io import loadmat
    import pandas as pd
    import numpy as np
    
    #loading mat file and create a dict object
    mat = loadmat(file)
    mat = {k:v for k, v in mat.items() if k[0] != '_'}
    
    df = pd.DataFrame()
    for i in range(len(feature_list)):
        #using flatten for converting to 1d array
        df[feature_list[i]] = mat[feature_list[i]].flatten()
    return df
```

```python
features = ['Mx', 'My', 'Mz', 'Pot1', 'Pot2']

data1 = read_matfile_to_pd('data/Pot1_0_Pot2_0_50_4X_wMag.mat',  features)
data2 = read_matfile_to_pd('data/Pot1_10_Pot2_0_50_4X_wMag.mat', features)
data3 = read_matfile_to_pd('data/Pot1_20_Pot2_0_50_4X_wMag.mat', features)
data4 = read_matfile_to_pd('data/Pot1_30_Pot2_0_50_4X_wMag.mat', features)
data5 = read_matfile_to_pd('data/Pot1_40_Pot2_0_50_4X_wMag.mat', features)
data6 = read_matfile_to_pd('data/Pot1_50_Pot2_0_50_4X_wMag.mat', features)
data7 = read_matfile_to_pd('data/Pot1_60_Pot2_0_50_4X_wMag.mat', features)

df = pd.concat([data1, data2, data3, data4, data5, data6, data7], ignore_index=True)
df
```

```python
df.to_csv('Combined_Data.csv')
```
